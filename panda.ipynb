```python
# Filter the data for Sunrisers Hyderabad wins by runs where Sunrisers Hyderabad can be in team1 or team2
sunrisers_wins_by_runs = data[
    (data['winner'] == 'Sunrisers Hyderabad') & 
    (data['win_by_runs'] > 0)
]

# Determine the opponent team using np.where
sunrisers_wins_by_runs['opponent'] = np.where(
    sunrisers_wins_by_runs['team1'] == 'Sunrisers Hyderabad', 
    sunrisers_wins_by_runs['team2'], 
    sunrisers_wins_by_runs['team1']
)

# Plot the wins by runs
plt.figure(figsize=(10, 6))
plt.bar(
    sunrisers_wins_by_runs['opponent'], 
    sunrisers_wins_by_runs['win_by_runs'], 
    color='orange', 
    edgecolor='black'
)
plt.xlabel('Opponent Team', fontsize=12)
plt.ylabel('Win by Runs', fontsize=12)
plt.title('Sunrisers Hyderabad Wins by Runs', fontsize=14)
plt.xticks(rotation=45, fontsize=10)
plt.show()
```

```python
delivarables=pd.read_csv('deliveries.csv')
delivarables.head()
```

```python
batsman=delivarables.groupby('batsman')
batsman.get_group('V Kohli').shape[0]

#ball played by kohli
```

```python
runs=batsman['batsman_runs'].sum().sort_values(ascending=False)
runs.head(5)
```

```python
#top most 6 by batsman
mask=delivarables['batsman_runs']==6
new_data=delivarables[mask]
new_data.shape[0]
new_data.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head(4)
```

```python
mask=delivarables['batsman']=='V Kohli'
new_data=delivarables[mask]
new_data.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(4)
```

```python
#most destructive batsman in deathover 16-20 acording to strike rate and 200+ balls faced
deathover=delivarables[delivarables['over']>15]
runs=deathover.groupby('batsman')['batsman_runs'].sum()
balls=deathover.groupby('batsman')['batsman_runs'].count()
strike_rate=(runs/balls)*100
mask=balls>200
strike_rate[mask].sort_values(ascending=False).head(5)
```

```python

matches=pd.read_csv('matches.csv')
new=delivarables.merge(matches,left_on='match_id',right_on='id')
new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season',keep='first').sort_values('season')
#seasonwise highest run scorer
```

```python
#Overwise runs scored by each team  
six=delivarables[delivarables['batsman_runs']==6]
ht=six.pivot_table(index=['over'],columns=['batting_team'],values='batsman_runs',aggfunc='count',fill_value=0)
```

```python




import seaborn as sns
from PIL import Image
sns.heatmap(ht, cmap='coolwarm')
```
